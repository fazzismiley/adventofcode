```python
with open('day6.txt', 'r') as f:
    lines = f.read().strip().splitlines()
print (len(lines))    
```

```python
#take the position of i append it to the dict when it does not exist and count the letters in that position appending
#to a nested dictionary by column position ie {pos: { letter: count}...}
#d = collections.Counter()
dict_count = {}
for line in lines:
    for i,val in enumerate(line):
        if i not in dict_count:
            dict_count[i] = {}
        elif i in dict_count:
            if val not in dict_count[i]:
                dict_count[i][val] = 1
            else:
                dict_count[i][val] += 1
```

```python
output = {}
for col, v in dict_count.items():
    tmp = {}
    for a, b in v.items():
        if not b in tmp:
            tmp[b] = [a]
        else:
            tmp[b].append(a)
        
    output[col] = tmp
```
